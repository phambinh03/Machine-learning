```python
#Importing dataset
# 1.1 read data 
filename = 'NHANES_age_prediction.csv'
df = pd.read_csv(filename)
df.head(10)
```

```python
names = list(df.columns)
print(names)
```

```python
types = df.dtypes
print(types)
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
for i in range(len(types)):
    if types[i]=='object':
        le.fit_transform(df[names[i]])
        df[names[i]] = le.transform(df[names[i]])

print('Successful')
```

```python
types = df.dtypes
print(types)
data = df.values
print(data[:10])
#pd.get_dummies(data=df)
```

```python
X = data[:,:-1]
y = data[:,-1]
print(X.shape)
print(y.shape)
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
print(X_scaled[0])
```

```python
from sklearn.preprocessing import Normalizer

norm = Normalizer()
df.iloc[:,1:-1] = norm.fit_transform(df.iloc[:,1:-1])
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split \
(X_scaled,y,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt


gnb = GaussianNB()
gnb.fit(X_train, y_train)
#gnb.paratical_fit(X_train,y_train)
y_pred = gnb.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print("MAE test score: ",mean_absolute_error(y_test,y_pred))
print("RMSE test score: ",sqrt(mean_squared_error(y_test,y_pred)))
```
